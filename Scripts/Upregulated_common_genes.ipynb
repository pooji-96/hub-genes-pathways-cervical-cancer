### CIN-N Group upregulated genes

```python
import pandas as pd

# File paths for CIN-N upregulated genes data
file_CIN_N = [
    "GSE63514/GSE63514_CIN-N_upregulated_gene_symbols.txt",
    "GSE64217/GSE64217_CIN-N_upregulated_gene_symbols.txt",
    "GSE138080/GSE138080_CIN-N_upregulated_gene_symbols.txt"
]

# Getting folder names
folder_names = [os.path.basename(os.path.dirname(path)) for path in file_CIN_N]
folder_names
```

```python
# Loading the data with folder name as column name
dfs_CIN_N = [pd.read_csv(path, header=None, names=[folder]) for path, folder in zip(file_CIN_N, folder_names)]
dfs_CIN_N[0].head()
```

```python
# Converting each column to a set of unique values
sets_CIN_N = [set(df[col]) for df, col in zip(dfs_CIN_N, folder_names)]

# Finding common genes
common_all_three_CIN_N = sets_CIN_N[0] & sets_CIN_N[1] & sets_CIN_N[2]
print(f"Common in all three folders: {len(common_all_three_CIN_N)}")
common_all_three_CIN_N
```

```python
# Finding the genes common in atleast two datasets only
common_at_least_two_CIN_N = (sets_CIN_N[0] & sets_CIN_N[1]) | (sets_CIN_N[0] & sets_CIN_N[2]) | (sets_CIN_N[1] & sets_CIN_N[2])
common_at_least_two_only_CIN_N = common_at_least_two_CIN_N - common_all_three_CIN_N
print(f"Only in exactly two folders (not all three): {len(common_at_least_two_only_CIN_N)}")
print(common_at_least_two_only_CIN_N)
```

### CC-N Group upregulated genes

```python
# File paths for CC-N upregulated genes data
file_CC_N = [
    "GSE63514/GSE63514_CC-N_upregulated_gene_symbols.txt",
    "GSE64217/GSE64217_CC-N_upregulated_gene_symbols.txt",
    "GSE138080/GSE138080_CC-N_upregulated_gene_symbols.txt"
]
```

### Venn diagram

```python
from venn import venn
import matplotlib.pyplot as plt

# Group titles and sets
group_titles = ['CIN-N', 'CC-CIN', 'CC-N']
group_sets = [sets_CIN_N, sets_CC_CIN, sets_CC_N]

# Creating a single figure with three subplots
fig, axes = plt.subplots(1, 3, figsize=(15, 5), dpi=300)

# Looping through each subplot and plotting the corresponding Venn diagram
for i, ax in enumerate(axes):
    data = {folder_names[j]: group_sets[i][j] for j in range(3)}
    
    # Plotting the Venn diagram on the corresponding axes
    venn(data, fmt="{size}", alpha=0.01, cmap=None, ax=ax, fontsize=10)
    
    ax.set_title(group_titles[i], fontsize=12)
    ax.axis('off')

# Saving the combined Venn diagram figure
plt.tight_layout()
plt.savefig("Upregulated_venn_combined.png", bbox_inches='tight', pad_inches=0.1)
plt.show()
```

### Common Upregulated genes

```python
all_upregulated = common_all_three_CIN_N | common_at_least_two_only_CIN_N | common_all_three_CC_CIN | common_at_least_two_only_CC_CIN | common_all_three_CC_N | common_at_least_two_only_CC_N
print(f"All upregulated genes: {len(all_upregulated)}")
print(all_upregulated)
```

```python
# Writing all upregulated genes to a .txt file
with open('all_upregulated.txt', 'w') as file:
    file.write('\n'.join(map(str, all_upregulated)))
```
